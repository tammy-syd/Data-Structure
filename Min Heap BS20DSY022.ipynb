```python
class MinHeap:
    def __init__(self):
        self.A = [0]
        self.size = 0
 
    # TC:O(1)
    # Return the position of
    # parent for the node currently
    # at i
    def parent(self, i):
        return i//2
 
    # TC:O(1)
    #  Return the position of
    # the left child of node at position i
    def left_child(self, i):
        return 2*i
 
    # TC:O(1)
    #  Return the position of
    # the right child of node at position i
    def right_child(self, i):
        return 2*i + 1
 
    # TC:O(logn)
    def insert(self, data):
        # Append  element to the heap
        self.A.append(data)
        # Increase size by 1
        self.size += 1
        # Move the element to its position from bottom to the top
        self.percolate_up(self.size)
 
    def percolate_up(self, i):
        # While the element is not the root or the left element
        while i//2 > 0:
            # If the element is less than its parent
            # swap the elements
            if self.A[i] < self.A[i//2]:
                # self.A[i//2], self.A[i] = self.A[i], self.A[i//2]
                temp = self.A[i]
                self.A[i] = self.A[i//2]
                self.A[i//2] = temp
            # Move the index to the parent to keep the properties
 
            i = i//2
 
    # TC:O(1)
    def get_minimum(self):
        if self.size == 0:
            return -1
        return self.A[1]
 
    # TC:O(logn)
    def deleteMin(self):
        # Get root of the heap
        result = self.A[1]
        # Move the last value of the heap to the root
        self.A[1] = self.A[self.size]
        self.size -= 1
        # Pop the last value
        self.A.pop()
        self.percolate_down(1)
        return result
 
    def percolate_down(self, i):
        # if the current node has at least one child
        while 2*i <= self.size:
            # Get the index of the min child of the current node
            min_child = self.minimum_child(i)
            # Swap the values of the current element is greater than its min child
            if self.A[i] > self.A[min_child]:
                self.A[i], self.A[min_child] = self.A[min_child], self.A[i]
            i = min_child
 
    def minimum_child(self, i):
        # If the current node has only one child 
        if 2*i+1 > self.size:
            # return the index of the unique child
            return 2*i
        else:
             # Return the index of the min child according to their values
            if self.A[2*i] < self.A[2*i+1]:
                return 2*i
            else:
                return 2*i+1
 
 
mh = MinHeap()
mh.insert(10)
mh.insert(8)
mh.insert(6)
mh.insert(4)
mh.insert(1)
mh.insert(5)
mh.insert(2)
mh.insert(0)
 
for i in range(8):
    print(mh.deleteMin())
```
